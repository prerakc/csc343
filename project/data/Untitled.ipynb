```python
carNames = pd.read_csv('CARS bkup\car-names.csv')
```

```python
display(carNames)
```

```python
modelList = pd.read_csv('CARS bkup\model-list.csv')
```

```python
display(modelList)
```

```python
combined = pd.merge(carNames, modelList, on='Model')
```

```python
display(combined)
```

```python
df = combined[['Id', 'ModelId', 'Make']].sort_values(by=['Id'])
```

```python
display(df)
```

```python
df.to_csv('CARS\car-names.csv', index=False)
```

```python
carsData = pd.read_csv('CARS bkup\cars-data.csv')
```

```python
display(carsData)
```

```python
carsDataCombined = pd.merge(carsData, df, on='Id')
```

```python
display(carsDataCombined)
```

```python
carsDataFiltered = carsDataCombined[['Id', 'MPG', 'Cylinders', 'Edispl', 'Horsepower', 'Weight', 'Accelerate', 'Year']]
```

```python
display(carsDataFiltered)
```

```python
carsDataFiltered.to_csv('CARS\cars-data.csv', index=False)
```
